###  Using Map with Panda DataFrame

```python
import os
import pandas as pd
import numpy as np

df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", 
    na_values=['NA', '?'])

display(df[0:10])
```

```python
list(df['origin'].unique())
```

```python
# make a new column with origin name
df['origin_name'] = df['origin'].map({
    1: 'North America',
    2: 'Europe',
    3: 'Asia'
})
```

```python
# shuffle the indices
df = df.reindex(np.random.permutation(df.index))
df.reset_index(inplace=True, drop=True)
df.head(10)
```

### Using Apply with Dataframe 

```python
effi = df.apply(lambda x: x['displacement']/ x['horsepower'], axis=1)
effi
```
